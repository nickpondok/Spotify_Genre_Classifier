The goal of this project is to classify genres based on spotify's audio features pulled from their API. As a part of my EDA process I did get rid of some genres from my classifier due to them being unusual genres such as 'at_home','happier_than_ever','in_the_car','wellness', etc. Below are boxplots for 3 differernt features.

I performed two baseline models using logistic regression (one vs rest) and KNN with a k-value of 115. Both models did not perform particularly well KNN had an accuracy of ~.28 and log reg had ~.31. The next steps would be to see if any feature engineering would help improve the models accuracy. I plan to focus more on logistic regression in order to get soft predictions. In my streamlit app I'd hope to return the top 3 potential genres since music can potentially fall into more than one genre. I will also need to weight my classes differently since there is some class imbalance.
